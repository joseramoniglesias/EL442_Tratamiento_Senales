# Carga de Imágen desde Dropbox

# Importación de Librerías

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# Lectura y Despliegue de Imagen

```python
img = cv2.imread('rice.png')
print('size = ',img.shape)
plt.imshow(img)
plt.show()
```

# Función de cómo es la imagen

```python
def howis(img):
  print('size = ',img.shape)
  print('max  = ',np.max(img))
  print('min  = ',np.min(img))

howis(img)
```

# Selección del primer canal de la imagen

```python
X = img[:,:,0]
howis(X)
plt.imshow(X)
```

# Función de Segmentación por Umbral

```python
def segmenta(X,t):
  (N,M) = X.shape
  Y = np.zeros((N,M))
  area = 0
  for i in range(N):
    for j in range(M):
      if X[i,j] > t:
        Y[i,j] = 255
        area = area + 1
  print('area = ',area)
  return Y
```

# Segmentación

```python
Y = segmenta(X,170)
plt.imshow(Y)
```

# Un ejemplo más complejo

```python
img = cv2.imread('rices.png')
X   = img[:,:,0]
howis(X)
plt.imshow(X)
```

```python
Y = segmenta(X,110)
plt.imshow(Y)
```

# Definición de imagen con fondo homogéneo

```python
(N,M) = X.shape
Xm    = np.zeros((N,M),np.uint8)
for i in range(N):
  xmin = np.min(X[i,:])
  Xm[i,:] = X[i,:] - xmin
plt.imshow(Xm)
```

```python
Y = segmenta(Xm,60)
plt.imshow(Y)
```
