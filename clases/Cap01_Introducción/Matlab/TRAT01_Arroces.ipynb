# Importación de Librerías

```python
import numpy as np
from cv2 import imread
from google.colab.patches import cv2_imshow
```

# Lectura y Despliegue de Imagen

```python
img = imread('onerice.png')
cv2_imshow(img)
```

# Función de cómo es la imagen

```python
def howis(img):
  print('size = ',img.shape)
  print('max  = ',np.max(img))
  print('min  = ',np.min(img))

howis(img)
```

# Selección del primer canal de la imagen

```python
X = img[:,:,0]
howis(X)
cv2_imshow(X)
```

# Función de Segmentación por Umbral

```python
def segmenta(X,t):
  (N,M) = X.shape
  Y = np.zeros((N,M))
  area = 0
  for i in range(N):
    for j in range(M):
      if X[i,j] > t:
        Y[i,j] = 255
        area = area + 1
  print('area = ',area)
  return Y
```

# Segmentación

```python
Y = segmenta(X,170)
cv2_imshow(Y)
```

# Un ejemplo más complejo

```python
!wget https://www.dropbox.com/s/n81hldnrcb07muh/rices.png
```

```python
img = imread('rices.png')
X   = img[:,:,0]
howis(X)
cv2_imshow(X)
```

```python
Y = segmenta(X,110)
cv2_imshow(Y)
```

# Definición de imagen con fondo homogéneo

```python
(N,M) = X.shape
Xm    = np.zeros((N,M),np.uint8)
for i in range(N):
  xmin = np.min(X[i,:])
  Xm[i,:] = X[i,:] - xmin
cv2_imshow(Xm)
```

```python
Y = segmenta(Xm,60)
cv2_imshow(Y)
```
