# Importación de Librerías

```python
import numpy as np
import cv2
from cv2 import imread
import matplotlib.pyplot as plt
#from google.colab.patches import cv2_imshow
```

# Lectura y Despliegue de Imagen

```python
img = imread('onerice.bmp')
plt.axis('off')
plt.imshow(img,cmap='gray')
```

# Función de cómo es la imagen

```python
def howis(img):
  print('size = ',img.shape)
  print('max  = ',np.max(img))
  print('min  = ',np.min(img))

howis(img)
```

# Selección del primer canal de la imagen

```python
X = img[:,:,0]
howis(X)
plt.axis('off')
plt.imshow(X,cmap='gray')
```

# Función de Segmentación por Umbral

```python
def segmenta(X,t):
  (N,M) = X.shape
  Y = np.zeros((N,M))
  area = 0
  for i in range(N):
    for j in range(M):
      if X[i,j] > t:
        Y[i,j] = 255
        area = area + 1
  print('area = ',area)
  return Y
```

# Segmentación

```python
Y = segmenta(X,170)
plt.axis('off')
plt.imshow(Y,cmap='gray')
```

# Definición de imagen con fondo homogéneo

```python
(N,M) = X.shape
Xm    = np.zeros((N,M),np.uint8)
for i in range(N):
  xmin = np.min(X[i,:])
  Xm[i,:] = X[i,:] - xmin
plt.axis('off')
plt.imshow(Xm,cmap='gray')
```

```python
Y = segmenta(Xm,60)
plt.axis('off')
plt.imshow(Y,cmap='gray')
```
