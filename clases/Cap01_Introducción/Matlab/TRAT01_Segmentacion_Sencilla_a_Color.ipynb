# Librerias

```python
import cv2
import numpy as np
from  cv2 import imread
import matplotlib.pyplot as plt
```

```python
def howis(img):
  print('size = ',img.shape)
  print('max  = ',np.max(img))
  print('min  = ',np.min(img))
```

# Lectura de imagen a color

```python
Icv = imread('flowers.jpg')               # Lectura en formato (BGR)
plt.axis('off')
plt.imshow(Icv)
howis(Icv)
```

```python
I = Icv[:,:,[2,1,0]]            # conversion a formato estndar (RGB)
plt.axis('off')
plt.imshow(I)
```

# Canales de color

```python
R = I[:,:,0]
G = I[:,:,1]
B = I[:,:,2]
RGB = np.concatenate((R,G,B),axis=1)
plt.axis('off')
plt.imshow(RGB,cmap='gray')
```

# Conversion a blanco y negro (tonos de gris)

```python
Rd = R.astype(float)
Gd = G.astype(float)
Bd = B.astype(float)
k  = (1/3,1/3,1/3)
Zd = k[0]*Rd+k[1]*Gd+k[2]*Bd
Z  = Zd.astype(int)
howis(Z)
plt.axis('off')
plt.imshow(Z,cmap='gray')
```

# Histograma

```python
def imhist(X):
  (N,M) = X.shape
  n = 256
  h = np.zeros((256,))
  for i in range(N):
    for j in range(M):
      x = X[i,j]
      h[x] = h[x]+1
  plt.plot(range(n),h[0:n])
  plt.show()
```

```python
imhist(Z)
```

# Segmentacin de la Flor Roja

```python
Sr = R>150
Sg = G<40
Sb = B<40 
Srgb = np.concatenate((Sr,Sg,Sb),axis=1)
plt.axis('off')
plt.imshow(Srgb*255,cmap='gray')
```

```python
S = np.logical_and(Sr,Sg,Sb)
plt.axis('off')
plt.imshow(S*255,cmap='gray')
```

# Eliminacin de filas con pocos unos

```python
(N,M) = S.shape
Q = S
for i in range(N):
  s = np.sum(S[i,:])
  if s<20:
    Q[i,:] = 0
plt.axis('off')
plt.imshow(Q*255,cmap='gray')
```

# Bsqueda de (imin,jmin) y (imax,jmax) en la regin segmetada

```python
imin = 1000
imax = 0
jmin = 1000
jmax = 0
for i in range(N):
  for j in range(M):
    if Q[i,j]>0:
      if i<imin:
        imin = i
      if i>imax:
        imax = i
      if j<jmin:
        jmin = j
      if j>jmax:
        jmax = j
```

# Deteccin de Bordes

```python
E = np.zeros((N,M),np.uint8)
for i in range(N):
  for j in range(1,M):
    if Q[i,j]!=Q[i,j-1]: 
      E[i,j]   = 1
      E[i,j-1] = 1
for i in range(1,N):
  for j in range(M):
    if Q[i-1,j]!=Q[i,j]: 
      E[i,j]   = 1
      E[i,j-1] = 1
plt.axis('off')
plt.imshow(E*255,cmap='gray')
```
