```python
import numpy as np
from cv2 import imread
import matplotlib.pyplot as plt
```

```python
def howis(img):
  print('size = ',img.shape)
  print('max  = ',np.max(img))
  print('min  = ',np.min(img))
```

```python
st_img = 'mandril.png'
#st_img = 'scale.png'
```

```python
img = imread(st_img)
X = img[:,:,0]
plt.imshow(X)
howis(X)
```

# Mustreo Espacial

```python
d = 4   # se muestrea cada d pixeles
(Nx,Mx) = X.shape
ix = range(0,Nx,d)
jx = range(0,Mx,d)
Ny = len(ix)
My = len(jx)
Y = np.zeros((Ny,My),np.uint8)
for i in range(Ny):
  for j in range(My):
    Y[i,j] = X[ix[i],jx[j]]
plt.imshow(Y)
howis(Y)
```

```python
d = 8  # se muestrea cada d pixeles
(Nx,Mx) = X.shape
Ny = Nx
My = Mx
Y = np.zeros((Ny,My),np.uint8)
for i in range(Ny):
  for j in range(My):
    ix = int(np.fix(i/d)*d)
    jx = int(np.fix(j/d)*d)
    Y[i,j] = X[ix,jx]
plt.imshow(Y)
howis(Y)
print('desplegado como = ',int(Nx/d),'x',int(Nx/d))
```

# Muestreo en tonos de gris (Cuantizacin)

```python
p = 4  # se muestrea el tono de gris cada p valores de gris
(Nx,Mx) = X.shape
Ny = Nx
My = Mx
Y = np.zeros((Ny,My),np.uint8)
for i in range(Ny):
  for j in range(My):
    x = int(np.fix(X[i,j]/p)*p)
    Y[i,j] = x
plt.imshow(Y)
howis(Y)
print('esta imagen tiene = ',int(256/p),' tonos de gris')
```

```python
import matplotlib.pyplot as plt
def imhist(X):
  (N,M) = X.shape
  n = 256
  h = np.zeros((256,))
  for i in range(N):
    for j in range(M):
      x = X[i,j]
      h[x] = h[x]+1
  plt.plot(range(n),h[0:n])
  plt.show()
```
