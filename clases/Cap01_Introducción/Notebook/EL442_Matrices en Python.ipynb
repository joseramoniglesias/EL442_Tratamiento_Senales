```python
from pylab import *
from skimage.color import rgb2gray
```

```python
A = imread('dataset/mona_lisa.jpg')
```

Con el código anterior hemos cargado una imagen y la hemos llevado a escala de grises para utilizar su matriz como ejemplo durante esta clase.

Para seleccionar los elementos en una matriz se necesitan dos índices: uno para establecer la localización en la fila y el otro para la columna correspondiente. En nuestro caso como trabajamos con imágenes cada elemento de la matriz le corresponde a un pixel.

<b>Obtener el valor de un pixel</b>

```python
A[2, 3]
```

<b>Obtener una columna específica con todas las filas</b>

```python
A[:, 3]
```

<b>Obtener una columna específica con los valores de las tres primeras filas</b>

```python
A[0:3, 3]
```

<b>De igual modo se puede aplicar para obtener una fila en un rango determinado de columnas</b>

```python
A[3,0:4]
```

<b>Utilizando este sistema de indexado podemos extrar regiones determinadas de la imagen (matriz)</b>

```python
imshow(A[80:105, 50:90])
```

Los indices que se utilizan en la matriz pueden ser:

* Un número entero comprendido entre $(0, n-1)$ done $n$ es el tamaño de la fila o columna

* Dos números separados por $:$ en la forma $x:y$ (rango) donde $x$ representa el índice inicial y $y$ el índice final.

* Tres números separados por $:$ en la forma $x:y:z$ (rango) donde $x$ representa el índice inicial y $y$ el índice final, y $z$ representa el "salto" o sea cuánto vamos sumando a $x$ hasta obtener $y$ o un valor mayor que $y$.


<b>En el siguiente ejemplo realizamos un recorrido por toda la imagen pero con un "salto" de 4</b>

Podemos ver como la imagen pierde resolución ya que se han descartado algunas filas y columnas

```python
imshow(A[::4,::4])
```

<b>También podemos utilizar saltos negativos</b>

```python
imshow(A[::-1, :])
```

Esto recorre la imagen en el eje $y$ a la inversa, la mimsa operación se puede realizar en el eje $x$

```python
imshow(A[:, ::-1])
```

<b>Además podemos utilizar indexado para obtener histogramas sobre una fila o columna determinada</b>

```python
plot(A[100,:])
```
